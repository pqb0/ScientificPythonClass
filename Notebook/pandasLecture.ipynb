```python
import pandas as pd

dictionary = {
    'Name' : ['Alice', 'Bob', 'Jeff', 'Chris'],
    'age' : [12, 22, 24, 70],
    'Occupation' : ['student', None, 'professor', None],
    '#Friends' : [None, 4.5, 18, 1]
}

df = pd.DataFrame(dictionary)
print(df)
print(df.head(1))

# reating a data frame
```

```python
#ADding Column

df['Salary'] = [None, None, 40000, 3200]
df
```

```python
#Adding Rows in 2 different ways

df.loc[len(df.index)] = ['Tom', 16, 'Dog Walker', 10, 350]


row = {'Name' : 'Jing', 'age' : 35, 'Occupation' : 'Student' , '#Friends' : 2, 'Salary': 0}
df = df._append(row, ignore_index=True)

#df1 = concat([df, df2, df3], ignore_index=true) To add multiple dataframes together.
df
```

```python
print(df) #Address specific value  for column and row

df.loc[df['Name'] == 'Alice', 'Salary'] = None
df
```
